```python
def answer_eleven():
    ContinentDict = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    
    global mergedData
    mergedData["Continent"] = [ContinentDict[x] for x in mergedData.index.tolist()]
    mergedData["Capita"] = mergedData["Energy Supply"].divide(mergedData["Energy Supply per Capita"])
    mergedGroupedData = mergedData[["Continent","Capita"]].groupby(by="Continent")
    
    funcs = [pd.core.groupby.DataFrameGroupBy.size,
             pd.core.groupby.DataFrameGroupBy.sum,
             pd.core.groupby.DataFrameGroupBy.mean,
             pd.core.groupby.DataFrameGroupBy.std]

    dfList = [y.to_frame(name=z)
              if type(y) == pd.core.series.Series
              else y.set_axis([z],axis="columns")
              for y,z in
                  [(x(mergedGroupedData),x.__name__)
                   for x in funcs]]
    return reduce(lambda x, y: pd.merge(x, y, left_index=True, right_index=True), dfList)
```

```python
def add_commas(num):
    regex = r"(\d*)(\.*)(\d*)"
    groups = re.match(regex,num)
    mSplit = [groups[1][::-1][start:start+3] for start in range(0, len(groups[1][::-1]), 3)]
    return ",".join(mSplit)[::-1] + groups[0].split(groups[1])[1]
    
mergedData["Capita"].astype(str).apply(add_commas)
```
