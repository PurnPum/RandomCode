```python
def prepareData(file,divisions=False,divdict=None,regexClean=None,fixNames=None):
    data=pd.read_csv(file)
    data2018 = data[data["year"] == 2018]
    if divisions and divdict is not None:
        dfDivs = [v.set_index(pd.MultiIndex.from_arrays([[k]*len(v),list(range(0,len(v)))])) for k,v in divdict.items()]
        data2018 = reduce(lambda x,y : x.append(y),dfDivs)
    
    if regexClean is not None:
        data2018["team"].replace(to_replace=regexClean,value='',regex=True,inplace=True)

    data2018WR = data2018[["team","W","L"]].copy()
    df_wr = data2018WR["W"].astype(int).divide(data2018WR["W"].astype(int) + data2018WR["L"].astype(int))
    df_wr.name = "WR"
    data2018WR = data2018WR.merge(df_wr,right_index=True,left_index=True)
    del data2018WR["L"]
    del data2018WR["W"]
    
    data2018WR["team"].replace(r'(.*)(\s)(\w*)$',r'\3',regex=True,inplace=True)
    if fixNames is not None:
        data2018WR["team"].replace(fixNames,inplace=True)
    
    return data2018WR
```

```python
def correlation(leagueName,dictLeague,dropRows,file,divisions=None,divdict=None,regexClean=None,fixNames=None): 
    
    cities = prepareCitiesLeagueData(leagueName,dictLeague,dropRows)
    data = prepareData(file,divisions,divdict,regexClean,fixNames)
    
    merged = data.merge(cities,left_on="team",right_on=leagueName)
    merged.drop(leagueName,axis=1,inplace=True)
    merged.set_index("Metropolitan area",inplace=True)
    
    wr_region = merged.groupby(by="Metropolitan area").mean().sort_values(by="Metropolitan area")
    population_region = merged["Population"].drop_duplicates().astype(float).to_frame().sort_values(by="Metropolitan area")

    population_by_region = population_region["Population"] # pass in metropolitan area population from cities
    win_loss_by_region = wr_region["WR"] # pass in win/loss ratio from nhl_df in the same order as cities["Metropolitan area"]

    assert len(population_by_region) == len(win_loss_by_region), "Q1: Your lists must be the same length"
    assert len(population_by_region) == 28, "Q1: There should be 28 teams being analysed for NHL"
    
    return stats.pearsonr(population_by_region, win_loss_by_region)
```

```python
NHL_wr_region = mergedNHL.groupby(by="Metropolitan area").mean().sort_values(by="Metropolitan area")
NBA_wr_region = mergedNBA.groupby(by="Metropolitan area").mean().sort_values(by="Metropolitan area")
MLB_wr_region = mergedMLB.groupby(by="Metropolitan area").mean().sort_values(by="Metropolitan area")
NFL_wr_region = mergedNFL.groupby(by="Metropolitan area").mean().sort_values(by="Metropolitan area")

dictWRs = {
    "NHL" : NHL_wr_region,
    "NBA" : NBA_wr_region,
    "MLB" : MLB_wr_region,
    "NFL" : NFL_wr_region
}
```

```python
sports = ['NFL', 'NBA', 'NHL', 'MLB']
p_values = pd.DataFrame({k:np.nan for k in sports}, index=sports)
c = combinations(sports,2)
```

```python
for a,b in c:
    m = dictWRs[a].merge(dictWRs[b],left_index=True,right_index=True,suffixes=('_'+a,'_'+b))
    p = stats.ttest_rel(m["WR_"+a],m["WR_"+b]).pvalue
    p_values.loc[a,b] = p
    p_values.loc[b,a] = p
```
